```python
F=q.drop(columns="y",axis=1)
T=q[["y"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test , y_train , y_test = train_test_split(F,T)
```

```python
from sklearn.preprocessing import MinMaxScaler
M=MinMaxScaler()
X_train[["pdays"]]=M.fit_transform(X_train[["pdays"]])
X_test[["pdays"]]=M.transform(X_test[["pdays"]])
```

```python
!pip install tensorflow
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
from tensorflow import keras
```

```python
model= Sequential([
    
    Dense(60,input_shape=(X_train.shape[1],),activation='relu'),#input layer
    Dense(30,activation='relu'),#hidden layer
    Dense(30,activation='relu'),
    Dense(30,activation='relu'),
    Dense(30,activation='relu'),#hidden layer
    Dense(1,activation='sigmoid'),#output layer
])
```

```python
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
model.fit(X_train,y_train,epochs=2)
```

```python
pred=model.predict(X_test)
pred
```

```python
pred1=[]
for i in pred:
    if i<=0.5:
        pred1.append(0)
    else:
        pred1.append(1)
```

```python
pred1
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix(y_test,pred1)
```

```python
accuracy_score(y_test,pred1)*100
```

```python
print(classification_report(y_test,pred1))
```
